### Zadanie 3

Celem tego zadania było sporządzenie wykresu funkcji f(x)=x^2+5, w odpowiednio zadanych zakresach, w tym celu należało wykorzystać bibliotekę numpy oraz matplotlib:

W pierwszym przypadku należało zademonstrować wykres funkcji na zakresie (-1,1) oraz odpowiednio opisać wykonany wykres.

```python

f = lambda x: x**2 + 5

x_1 = np.linspace(-1, 1)
y_1 = f(x_1)

plt.plot(x_1, y_1, label = 'x^2+5')
plt.title('Wykres funkcji x^2 + 5 dla zakresu (-1,1)')
plt.legend()
plt.xlabel('Argumenty')
plt.ylabel('Wartości funkcji')
```

W tym przypadku należało postąpić analogicznie tym razem dla zakresu (-6,6).

```python
x_2 = np.linspace(-6,6)
y_2 = f(x_2)

plt.plot(x_2, y_2, label = 'x^2+5')
plt.title('Wykres funkcji x^2 + 5 dla zakresu (-6,6)')
plt.legend()
plt.xlabel('Argumenty')
plt.ylabel('Wartości funkcji')
```

W ostatnim podpunkcie również dla tej samej funkcji, należy przedstawić wykres na zakresie (0,5).

```python
x_3 = np.linspace(0, 5)
y_3 = f(x_3)

plt.plot(x_3, y_3, label = 'x^2+5')
plt.title('Wykres funkcji x^2 + 5 dla zakresu (0,5)')
plt.legend()
plt.xlabel('Argumenty')
plt.ylabel('Wartości funkcji')
```
